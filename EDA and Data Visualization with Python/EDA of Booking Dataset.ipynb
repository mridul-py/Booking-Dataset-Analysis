# Popularity of Time Slots

## Convert "Time Slot" to Categorical Format
- **Purpose:** To transform the 'Time Slot' column (assumed to be in HH:MM format) into a categorical format by extracting only the hour.
- **Details:** This step creates a more uniform representation of time slots, making it easier to group and analyze data based on hourly trends.

## Count Bookings Per Time Slot
- **Purpose:** To aggregate booking counts by each hourly time slot.
- **Details:** The grouped data provides insight into the frequency of bookings during different hours, helping to identify peak activity times.

## Prepare Heatmap Data
- **Purpose:** To restructure the data into a heatmap-friendly format.
- **Details:** Pivoting the data simplifies visualization by organizing it into rows and columns suitable for a heatmap.

## Plot Heatmap
- **Purpose:** To visually analyze the popularity of different time slots.
- **Details:** The heatmap shows booking density using color gradients, making patterns easy to spot (e.g., hours with the highest or lowest bookings).
- **Analysis Tip:** Use this visualization to identify trends in booking behaviors, such as peak or quiet hours, and adjust strategies accordingly.

````python
# Convert "Time Slot" to a categorical format (assuming it's in HH:MM format)
df['Time Slot'] = pd.to_datetime(df['Time Slot'], format='%H:%M').dt.hour  # Extracting hour

# Count bookings per time slot
time_slot_counts = df.groupby('Time Slot').size().reset_index(name='Booking Count')

# Pivot to create heatmap format (if needed)
heatmap_data = time_slot_counts.pivot_table(values='Booking Count', index='Time Slot')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Popularity of Time Slots")
plt.xlabel("Time Slot (Hour)")
plt.ylabel("Number of Bookings")
plt.show()
````

# Handling Outliers in the "Price" Column

## Calculate Quartiles and Interquartile Range (IQR)
- **Purpose:** To determine the spread of the data in the "Price" column by calculating the first quartile (Q1), third quartile (Q3), and the interquartile range (IQR).
- **Details:** These metrics help define the range within which most data points lie.

## Define Bounds for Outlier Detection
- **Purpose:** To establish thresholds for identifying outliers using the formula:
  - **Lower Bound:** Q1 - 1.5 * IQR
  - **Upper Bound:** Q3 + 1.5 * IQR
- **Details:** Data points outside these bounds are considered outliers.

## Filter Out Outliers
- **Purpose:** To remove outliers from the dataset by retaining only the rows where "Price" falls within the calculated bounds.
- **Details:** This step ensures cleaner data and reduces distortion caused by extreme values.

## Compare Dataset Size
- **Purpose:** To evaluate the impact of outlier removal by comparing the original dataset size with the size of the filtered dataset.
- **Details:** It quantifies the number of rows removed during this process, providing insights into how prevalent outliers are in the "Price" column.

````python
Q1 = df["Price"].quantile(0.25)  # First quartile
Q3 = df["Price"].quantile(0.75)  # Third quartile
IQR = Q3 - Q1  # Interquartile range

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_no_outliers = df[(df["Price"] >= lower_bound) & (df["Price"] <= upper_bound)]

# Check the difference
print(f"Original dataset size: {df.shape[0]}")
print(f"After removing outliers: {df_no_outliers.shape[0]}")

````

## Histograms for All Numerical Columns
- **Purpose:** To explore the frequency distribution of numerical data across all columns.
- **Details:** Histograms provide a quick overview of value frequencies within specific ranges, helping identify patterns such as skewness, gaps, or multimodal distributions.

## KDE Plots for Distribution Shapes
- **Purpose:** To visualize the probability density and distribution shape of each numerical column using Kernel Density Estimation (KDE).
- **Details:** KDE plots offer a smooth curve that represents the data distribution, making it easier to observe overall patterns and peaks compared to histograms.
- **Analysis Tip:** Pay close attention to the shape of the curve (e.g., symmetry, peaks, or tails), as it provides insight into the data structure and informs potential transformations.

````python
# Histograms for all numerical columns
df.hist(figsize=(12, 8), bins=30)
plt.show()

# KDE plots to see the distribution shape
for col in df.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(df[col], fill=True)
    plt.title(f'Distribution of {col}')
    plt.show()
````
